```python
import pandas as pd
import numpy as np
import scipy as sp
```

```python
file = open('train.dat')
lst = []
for line in file:
    lst+=[line.split('\t')]

#test the result
print lst[0]
print len(lst)
```

```python
#put into dataFrame
dataf = pd.DataFrame(lst)
dataf.columns = ['res','doc']
dataf.head()
```

```python
docs = dataf.doc
res = dataf.res
```

```python
docs.shape
```

```python
# Train on a 80/20 split
from sklearn.model_selection import train_test_split
X_train, X_test,y_train,y_test = train_test_split(docs,res, test_size=0.2, random_state =1)
```

```python
len(X_train),len(y_train)
```

```python
y_train.value_counts()
```

```python
#convert into a matrix of token counts
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(lowercase     = True ,
                             analyzer      = 'word',
                             stop_words    = 'english')
```

```python
vectorizer.fit(X_train)
#vectorizer.get_feature_names()
```

```python
X_train_dtm = vectorizer.transform(X_train)
#pd.DataFrame(X_train_dtm.toarray(),columns=vectorizer.get_feature_names())
```

```python
X_test_dtm = vectorizer.transform(X_test)
#print(X_test_dtm)
#pd.DataFrame(X_train_dtm.toarray(),columns=vectorizer.get_feature_names())
```

```python
#idf the docs
```

```python
#l2-norm (make the length of the vector to 1) for easy calculation for cosine similarity
```

```python
from sklearn.preprocessing import Normalizer
normalizerl2 = Normalizer(norm = "l2")
normalizerl2.fit(X_train_dtm)
X_train_dtm_norml2 = normalizerl2.transform(X_train_dtm)

```

```python
print(X_train_dtm_norml2)
```

```python
#l2-norm the test set
```

```python
X_test_dtm_norml2 = normalizerl2.transform(X_test_dtm)
print(X_test_dtm_norml2)
```

```python
# compute the similary of the test set data to all the data in the training set
# save the pairwise similary into a similarity matrix for later knn use
```

```python
from sklearn.metrics.pairwise import linear_kernel
cosine_similarities = linear_kernel(X_test_dtm_norml2[1:2], X_train_dtm_norml2).flatten()
print (cosine_similarities)
```

```python
related_docs_indices = cosine_similarities.argsort()[:-10:-1]
print related_docs_indices
cosine_similarities[related_docs_indices]
```
